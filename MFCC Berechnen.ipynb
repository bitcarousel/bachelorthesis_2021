```python
# Load Bibliotheken
import os
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
```

```python
# Importieren des Hallo min Name isch Christian File
audio_file = "audio/Hallo.wav"
```

```python
# Abspielen der Audiodatei
ipd.Audio(audio_file)
```

```python
# Laden des Audio Files
scale, sr = librosa.load(audio_file)
```

# Extrahieren der MFCC's

```python
# Filterbanks auf 10 minimieren. 
filter_banks = librosa.filters.mel(n_fft=2048, sr=22050, n_mels=10)
```

```python
# Filter Banks Shape überprüfen
filter_banks.shape
```

```python
# Mel Spektrogram erstellen
mel_spectrogram = librosa.feature.melspectrogram(scale, sr=sr, n_fft=2048, hop_length=512, n_mels=10)
```

```python
# Mel Spektrogram Shape überprüfen 
mel_spectrogram.shape
```

```python
# Log Mel für bessere Darstellung
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
```

```python
# Log Mel Spektrogram Shape überprüfen
log_mel_spectrogram.shape
```

# Visualisierung der MFCC's

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(log_mel_spectrogram, 
                         x_axis="time",
                         y_axis="mel", 
                         sr=sr)
cbar = plt.colorbar(format="%+2.f")
plt.xlabel('Zeit in (s) ', fontsize=20)
plt.ylabel('Frequenz in (Hz)', fontsize=20)
cbar.set_label('Anzahl Energie in (dB)', fontsize=20)
cbar.ax.tick_params(labelsize=15) 
plt.xticks(size = 20)
plt.yticks(size = 20)
plt.show()
```
